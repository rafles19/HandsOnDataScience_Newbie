```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("/content/drive/MyDrive/Buku-Machine-Learning-main/Data Normalization and Discretization/datasetgaji2.csv")
```

```python
df.head()
```

Simple Feature Scaling

rumus : *X_new = X / X_max*


```python
xmax =  df['Umur'].max()
xmax
```

```python
df['simple_scaling'] = df['Umur'] / xmax
df
```

Min-Max *Normalization*

rumus X = (x-x_min)/(x_max-x_min)

```python
xmin = df['Umur'].min()
xmin
```

```python
df['minmax'] = (df['Umur'] - xmin) / (xmax - xmin)
```

```python
print(df['minmax'])
```

```python
#menggunakan library sklearn
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['Umur']])
print(scaled)
```

*Mean Normalization*


*   rumus1 : Xavg = jumlah total x / jumlah data
*   rumus2 : X = (x - Xavg) / (x_max-x_min)



```python
xavg = df['Umur'].mean()
xavg
```

```python
df['Mean Norm'] = (df['Umur'] - xavg)/(xmax-xmin)
df
```

Standarization Z-Score



```python
from sklearn.preprocessing import StandardScaler
scaler_z = StandardScaler()
```

```python
scaled_z = scaler_z.fit_transform(df[['Umur']])
print(scaled_z)
```
